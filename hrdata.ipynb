
<h3>Total Employees by gender</h3>


```python
ax=sns.countplot(x='gender',data=df)
plt.title('Total Employee By Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
employees_left=sns.countplot(x='attrition',data=df)
plt.title('Attrition Count')
for bars in employees_left.containers:
    employees_left.bar_label(bars)
    
```


<h3>Attrition Count by Gender</h3>


```python
attrition_by_gender = df.groupby(['gender', 'attrition_label']).size().unstack(fill_value=0)
print(attrition_by_gender)
```

```python
ax = attrition_by_gender.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')

# Add labels
for container in ax.containers:
    ax.bar_label(container)

```


<h3> Department Wise Attrition </h3>


```python
department_wise_attrition = df.groupby(['department', 'attrition_label']).size().unstack(fill_value=0)
print(department_wise_attrition)
```

```python
ax = department_wise_attrition.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('department_wise_attrition ')
plt.xlabel('department')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
for container in ax.containers:
    ax.bar_label(container)

```


<h3>Education Field Wise Attrition </h3>


```python
education_field_wise_attrition = df.groupby(['education_field', 'attrition_label']).size().unstack(fill_value=0)
print(education_field_wise_attrition)
```

```python
ax = education_field_wise_attrition.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title(' education_field_wise_attrition')
plt.xlabel('education_field')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
for container in ax.containers:
    ax.bar_label(container)
```


<h2>Age Band</h2>

```python
ax=sns.countplot(x='age_band',data=df)
plt.title(' Age Group Wise Employee include Attrition')
for bars in ax.containers:
    ax.bar_label(bars)
```
